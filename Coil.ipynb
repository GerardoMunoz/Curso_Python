Next, I graphed the magnetic field as a function of distance and created a log-log plot of the data. The log-log plot exhibited a linear relationship, confirming that the magnetic field follows a power-law decay with distance.





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
distance_cm = np.array([22, 21, 20, 19, 18, 17, 16, 15, 14, 13, 12, 11, 10, 9, 8, 7, 6, 5, 4])
magnetic_field_uT = np.array([57, 61, 66, 73, 81, 90, 102, 116, 134, 157, 186, 223, 277, 358, 461, 601, 814, 1197, 1792])
log_field = np.log10(magnetic_field_uT)
log_distance = np.log10(distance_cm)

# Perform linear regression on log-log data
slope, intercept, r_value, p_value, std_err = linregress(log_distance, log_field)

# Plot log-log relationship
plt.figure(figsize=(8, 6))
plt.scatter(log_distance, log_field, label="Data Points", color="blue")
plt.plot(log_distance, slope * log_distance + intercept, label=f"Fit: Slope = {slope:.2f}", color="red")
plt.xlabel("Log(Distance) [log(cm)]")
plt.ylabel("Log(Magnetic Field) [log(µT)]")
plt.title("Log-Log Plot of Magnetic Field vs. Distance")
plt.legend()
plt.grid(True)
plt.show()

# Regression results
slope, intercept, r_value**2
```

Using this linear trend, I derived the equation that represents the relationship between the magnetic field and distance.

## Coil

The magnetic field $B$ along the axis of a circular coil at a distance $d$ from its center is given by the formula:

$B = \frac{\mu_0 \, N \, I \, R^2}{2 \, (R^2 + d^2)^{3/2}}$

Where:
* $\mu$ : The permeability of free space ($4 \pi \times 10 ^{-7} T \cdot m / A$)
* $N$: Number of turns in the coil.
* $I$: Current passing through the coil (in amperes).
* $R$: Radius of the coil (in meters).
* $d$: Distance from the center of the coil along its axis (in meters).

```python
import math
mu=4*math.pi*10**(-7)
N=133
I=1
R=0.005
d=0.01
B=4e-7*math.pi*N*1*R**2/(2*(R**2+d**2)**(3/2))
print(f"B={B}")
```
